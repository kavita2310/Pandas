1. Add, subtract, multiple and divide two series of Pandas.

Series1: [2, 4, 6, 8, 10]
series2: [1,2,3,4,5]

```python
import pandas as pd
s1=pd.Series([2, 4, 6, 8, 10])
s2=pd.Series([1,2,3,4,5])
print(s1+s2)
print(s1-s2)
print(s1*s2)
print(s1/s2)
```

20. Write a Pandas program to join (left join) the two dataframes using keys from left dataframe only.

Data:
    
df1: 
Year    Week    Colour    Val1 
2020       A       Red      50
2020       B       Red      60
2020       B     Black      70
2020       C       Red      10
2020       D     Green      20

df2:
Year    Week    Colour    Val2
2020       A     Black      30
2020       B     Black     100
2020       C     Green      50
2020       C       Red      20
2020       D       Red      40



```python
df1=pd.DataFrame({'Year':[2020,2020,2020,2020,2020],'Week':['A','B','B','C','D'],
'Colour':['Red','Red','Black','Red','Green'],'Value':[50,60,70,10,20]})
df1
```

```python
df2=pd.DataFrame({'Year':[2020,2020,2020,2020,2020],'Week':['A','B','C','C','D'],
'Colour':['Black','Black','Green','Red','Red'],'Value':[30,100,50,20,40]})
df2
```

```python
pd.merge(df1,df2,how='left',on=['Week','Colour'])
```
